```python
def plot_orderbook_shape(orderbook_shape_df):
    ax = plt.gca()

    # Move left y-axis and bottim x-axis to centre, passing through (0,0)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')

    # Eliminate upper and right axes
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    # Show ticks in the left and lower axes only
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    ax.plot(orderbook_shape_df['cum_dollar'],orderbook_shape_df['price_change_from_mq'],marker='*')
    plt.xlabel('cumulative depth',horizontalalignment='right', x=1.0)
    plt.ylabel('relative price deviation from midpoint',horizontalalignment='right', y=1,rotation='horizontal')
```

```python
plot_orderbook_shape(orderbook_shape20180101)
```

```python
plot_orderbook_shape(orderbook_shape20181001)
```
