```python
from liquidity_metrics import *
import matplotlib
%matplotlib notebook
#%matplotlib inline
```

```python
dates = [20180101, 20181001]
```

```python
data = {}
for this_date in dates:
    quote = load_quote(this_date)
    depth = load_depth(this_date)
    trade = load_trade(this_date)
    data[this_date] = {'quote':quote,
                       'trade':trade,
                       'depth':depth}
    print(this_date, 'loading done')
```

```python
for k,v in data[20180101].items():
    print(k,'head: ')
    print(v.head())
    print('\n')
```

```python
def plot_hourly_dollar_volume(trade_of_date,*args,**kwargs):
    hourly_dollar_volume = trade_of_date.groupby(trade_of_date['time_exchange'].dt.hour).apply(dollar_volume)
    plt.plot(hourly_dollar_volume, *args, **kwargs)
```

```python
fig, axes = plt.subplots(len(dates), 1,sharex=True,sharey=True)
for idx,this_date in enumerate(dates):
    trade = data[this_date]['trade']
    plt.sca(axes[idx])
    plot_hourly_dollar_volume(trade,marker='*')
    axes[idx].set_title(this_date)
    plt.subplots_adjust(top=2.)
```

```python
def plot_hourly_meanbaspread(depth_of_date,*args,**kwargs):
    hourly_meanbaspread = depth_of_date.groupby(depth_of_date['time_exchange'].dt.hour).apply(mean_baspread)
    plt.plot(hourly_meanbaspread, *args, **kwargs)
```

```python
fig, axes = plt.subplots(len(dates), 1,sharex=True,sharey=True)
for idx,this_date in enumerate(dates):
    depth = data[this_date]['depth']
    plt.sca(axes[idx])
    plot_hourly_meanbaspread(depth,marker='*')
    axes[idx].set_title(this_date)
    plt.subplots_adjust(top=2.)
```

```python
def plot_hourly_basbp(depth_of_date,*args,**kwargs):
    hourly_basbp = depth_of_date.groupby(depth_of_date['time_exchange'].dt.hour).apply(mean_baspread_basispoint)
    plt.plot(hourly_basbp, *args, **kwargs)
```

```python
fig, axes = plt.subplots(len(dates), 1,sharex=True,sharey=True)
for idx,this_date in enumerate(dates):
    depth = data[this_date]['depth']
    plt.sca(axes[idx])
    plot_hourly_basbp(depth,marker='*')
    axes[idx].set_title(this_date)
    plt.subplots_adjust(top=2.)
```

```python
def plot_hourly_depth(depth_of_date,*args,**kwargs):
    hourly_depth = depth_of_date.groupby(depth_of_date['time_exchange'].dt.hour).apply(market_depth).unstack()
    plot_market_depth(hourly_depth.loc[:,'bid'],hourly_depth.loc[:,'ask'])
```

```python
for idx,this_date in enumerate(dates):
    depth = data[this_date]['depth']
    plot_hourly_depth(depth,marker='*')
    plt.suptitle('dollar depth at {}'.format(this_date))
```

```python
def plot_hourly_change_percent(depth_of_date,*args,**kwargs):
    hourly_prcpct = depth_of_date.groupby(depth_of_date['time_exchange'].dt.hour).apply(price_change_percent).unstack()
    plot_prcpct(hourly_prcpct.loc[:,'bid'],hourly_prcpct.loc[:,'ask'])
```

```python
for idx,this_date in enumerate(dates):
    depth = data[this_date]['depth']
    plot_hourly_change_percent(depth,marker='*')
    plt.suptitle('hourly price deviation at {}'.format(this_date))
```

```python
def plot_orderbook_shape(orderbook_shape_df):
    ax = plt.gca()

    # Move left y-axis and bottim x-axis to centre, passing through (0,0)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')

    # Eliminate upper and right axes
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    # Show ticks in the left and lower axes only
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    ax.plot(orderbook_shape_df['cum_dollar'],orderbook_shape_df['price_change_from_mq'],marker='*')
    plt.xlabel('cumulative depth',horizontalalignment='right', x=1.0)
    plt.ylabel('relative price deviation from midpoint',horizontalalignment='right', y=1,rotation='horizontal')
```

```python
fig, axes = plt.subplots(len(dates), 1,sharex=True,sharey=True)
for idx,this_date in enumerate(dates):
    depth = data[this_date]['depth']
    plt.sca(axes[idx])
    plot_orderbook_shape(shape_of_order_book(data[this_date]['depth']))
    axes[idx].set_title('ave orderbook shape at {}'.format(this_date),loc='right')
    plt.subplots_adjust(top=2.)
```

```python
def save_hourly_orderbook_shape(this_date, depth_of_date):
    hourly_shape = depth_of_date.groupby(depth_of_date['time_exchange'].dt.hour).apply(shape_of_order_book)
    import matplotlib.animation as animation
    fig = plt.figure(figsize=(8, 6), dpi=80)
    def draw_each_frame(index):
        plt.cla()
        plt.title("shape of order book at time {} {}:00".format(this_date, index),y=-0.1)
        plt.grid(True)
        plt.xlim(-150000,150000) #hard-coded
        plt.ylim(-0.15,0.15)     #hard_coded
        plot_orderbook_shape(hourly_shape.loc[index])

    anim = animation.FuncAnimation(fig, draw_each_frame, 
                                   frames=24, interval=200)
    anim.save('shape_of_orderbook_each_hour_{}.gif'.format(this_date),writer='imagemagick')
```

```python
for idx,this_date in enumerate(dates):
    save_hourly_orderbook_shape(this_date, data[this_date]['depth'])
```
